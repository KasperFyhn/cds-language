Q: Where and when to create variables, assign values and alter them in (nested) for-loops?

```python
subfolder_to_text_lengths = {}

# outer loop for subfolders in the main folder
subfolders = glob('USEcorpus/*')
for subfolder in subfolders:

    # create a list for subfolder to which I will append stats from individual files
    text_lengths = []

    # inner loop for files in the given subfolder (changes each iteration)
    files_in_subfolder = glob(subfolder + '/*')

    for file in files_in_subfolder:
        # process individual file
        with open(file, encoding='latin-1') as f:
            text = f.read()
            # remove id tags
            # skipping the actual processing
            # doc = nlp(text)
        print('Simulating processing of', file)
        text_length = len(text)
        text_lengths.append(text_length)

    # print(subfolder, text_lengths)
    subfolder_to_text_lengths[subfolder] = text_lengths

print(subfolder_to_text_lengths)
```
